# #Patika Temel Matematik Dersi Bitirme Projesi:

Herkese merhabalar. Yukarıdaki başlıkta da görebileceğiniz gibi ilgili dersin bitirme projesi için aşağıda basit bir matematik sorusunun sezgisel yolla çözümünü yapmaya çalışacağım. Açıklamalarım umarım sizlere de bir anlam ifade eder, benzer bir şeyler yapmaya çalışırken sizin de işinize yarar. 

Evet sorumuza geçelim:

Elimizde tek bilinmeyenli şöyle bir denklem olsun:

(x+7)/3+(x+4)/4+x/7+(x+1)/9+(x+8)/10+22=0

x=?

Şimdi, yukarıdaki soruyu bildiğimiz temel matematiksel yöntem ile (hamaliyesi en ağır olan metot ile) şöyle çözebiliriz: İlkokul bilgisi ile paydaları eşitleyerek kesirli ifadeleri paydadan kurtarırız. Sonra x’ i  denklemin bir tarafında yalnız bırakarak sonuca ulaşırız. Paydaları eşitlemek problem değil de, fakat maşallah paydadaki sayılar da -neredeyse tamamen- asal sayı olduğundan, paydadaki sayıların en küçük ortak katı oldukça büyük bir sayı çıkacaktır haliyle -> 

OKEK(3,4,7,9,10)=1260

Bu yüzden de matematiksel işlem yükümüz çok artacaktır. 

Örneğin sadece denklemdeki ilk eleman için düşünecek olsak bile paydayı 1260'a eşitlemek için ilk elemanı 420 ile çarpacağız. Böylelikle ilk eleman:

420*(x+7)/1260 = (420x + 2940)/1260 olacak.

Problemdeki tüm elemanlar için bunu tekrarladığımızı düşünecek olursak bir hayli zaman kaybedeceğimiz aşikar. Zaten ben buraya kadar yazarken bile yoruldum. Nihayetinde denklem şöyle bir şeye dönüşecek: 

((420x+2940))/1260+(315*(x+4))/1260+180x/1260+(140*(x+1))/1260+(126*(x+8))/1260+22*1260=0
Sizin de okurken içiniz şişmesin diye burada kesiyorum...

Buraya kadar ağır işçilik sergilememiz gereken çözümü kısmen gördünüz. İlkokul matematik bilgisine güvenerek gerisinin nasıl getirileceğini tahmin etmek zor değil. Buradan sonra ise konunun da esas içeriği olan "sezgisel çözüm" hakkında yazmak istiyorum. Bu soruyu dakikalarca elle hesap yaparak debelenmeden nasıl çözeriz?

"Sezgisel Metot":

Tekrar başa dönelim. Denklem şöyle bir şey idi:

(x+7)/3+(x+4)/4+x/7+(x+1)/9+(x+8)/10+22=0

Şimdi probleme baktığımızda eşitlikten hemen önce en sonda bir "+22" ifadesi görüyoruz. Soldaki kesirli ifadelerin hepsinin toplamı belli ki bir negatif sayı olmalı. Yoksa x içeren kesirli ifadeler ancak 22 ile toplandığında bize sıfırı vermezdi. Böylelikle ilk "sezmemiz" gereken x'in negatif bir değer olması gerektiği. Bakın denklemdeki tüm elemanlar negatif sonuç vermeyebilir. Örneğin x/7 negatif olabilir ama x+8/10 pozitif de olabilir. Söylemek istediğim x’in bir negatif değer taşıması gerektiği. Şimdi denklemdeki herhangi iki elemanı alalım. Mesela burada:

(x+4)/4  ve  x/7  'yi

Neden bunları seçtik? Paydalar birbiri ile alakasız sayılar, ortak böleni yok. Dolayısıyla ortak en küçük katları da birbirlerinin çarpımı oluyor. Böyle bir sayı üzerinde de deneme yanılma yapmak daha kolay olacak. Yalnız önemli husus sayının negatif olması. Onun da sebebini yukarıda konuştuk zaten. Şimdi eğer x’i -28 alırsak, bu yukarıda seçtiğim ikili neye dönüşür?

(-28+4)/4=-6 ve  (-28)/7=-4 güzel, iki tane tamsayımız var. Belki doğru yoldayızdır, bakalım…

Peki bu x değerini diğer elemanlar için de kullanırsak tamsayı sonuçları alabiliyor muyuz? Paydası 3, 9 ve 10’lu olan ifadeler için bakalım:

(-28+7)/3=-7      (-28+1)/9=-3      (-28+8)/10=-2

Güzel, hepsi tamsayılı sonuçlar verdiler. Şimdi bunları denklemde yerlerine koyalım:

-7-6-4-3-2=-22 => -22+22=0

Demek ki yolumuz doğru ve x=-28 imiş. 

Peki başka açıdan düşünelim. Neden ilk başta 7 ve 4 paydasına sahip olan sayıları seçtik? Tamam, ortak en küçük katları 7*4=28 ama bunun gibi düşünebileceğimiz ve hiçbir ortak bölenleri olmayan başka sayı çiftleri de yok mu? Var, misal: (4,9), (3,7) veya (3,10). Doğru haklısınız, ama kısa bir kontrolle bunların denklemde çok da işimize yarayan sonuçlar vermediğini görebiliriz. 

Şöyle ki:
(4,9) İkilisini ele alırsak:
4*9=36 ve x=-36 dersek ifadeler sırasıyla:

(-36+4)/4=-8           (-36+1)/9=>(-35)/8=-3.88889

(3,7) İkilisini ele alırsak:
3*7=21 ve x=-21 dersek ifadeler sırasıyla:

(-21+4)/4=-4.25      (-21)/7=-3

Ya da bir diğer alternatif olan (3,10) İkilisini ele alırsak:
(-30+7)/3=7.66667      (-30+8)/10=-2.2 görülebileceği gibi işler daha da karışıyor.

Bu tarz sonuçlar diğer ikililerde de benzer çıktığından bir iki deneme sonucunda en makul çözümün -28 olduğunu bulabiliriz. Tüm bu denklemi tek tek elle hesaplayarak çözmek yerine bu tarz bir yaklaşım hem daha hızlı hem de daha kolay şekilde sonuca ulaşmamızı sağlıyor. Ben bu tarz bir çözüme sezgisellikle yaklaşılan çözüm diyorum, siz isterseniz mahmut da diyebilirsiniz…

Okuyanlara selamlar…
